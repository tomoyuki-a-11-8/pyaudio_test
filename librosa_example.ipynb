```python
import librosa
import matplotlib.pyplot as plt
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
list_ceps = []  # 抽出したMFCCを格納するリスト
list_label = []  # 正常(0)、異常(1)ラベルを格納するリスト
```

```python
# 440のwavを処理
filelist = glob.glob('test_wave/440*.wav')  # ワイルドカードでnormal*.wavのリストを作成

for filename in filelist:
    y, sr = librosa.core.load(filename, sr=22050, duration = 10)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20,) #sr:sampling rate is changed to 44,100Hz Default setting is 22.050Hz
    
    # 複数のローリングウィンドウでそれぞれ20次元のMFCCを得られるので、その平均をとる。
    ceps = mfcc.mean(axis=1)
    
    # リストに追加
    list_ceps.append(ceps)  # 20次元のMFCCを追加
    list_label.append(0)  # 正常ラベル(0)を追加
```

```python
pd.DataFrame(mfcc)
```

```python
pd.DataFrame(ceps)
```

```python
import librosa.display
librosa.display.specshow(mfcc, x_axis='time')
```
